# Introducción
En este tutorial, exploraremos conceptos básicos de probabilidad y estadística en R, utilizando el dataset `iris`. Aprenderemos sobre generación de números aleatorios,  resumen estadístico de datos y distribuciones de probabilidad.

```r
library(tidyverse)
library(ggplot2)
library(dplyr)
```

# Números aleatorios y probabilidad

La generación de números aleatorios es fundamental en estadística y probabilidad. Podemos generar números aleatorios de diferentes distribuciones y calcular probabilidades.

```r
print(runif(10, min = 0, max = 1))
print(rnorm(10, mean = 0, sd = 1))
print(pbinom(3, size = 10, prob = 0.5))
```

# Exploración del dataset `iris`

El dataset `iris` contiene mediciones de diferentes características de flores de tres especies. Lo utilizaremos para calcular estadísticas descriptivas y visualizar distribuciones.

```r
data(iris)
```

```r
head(iris) # Muestra las primeras filas
dim(iris)  # Dimensión del dataset
summary(iris)  # Resumen estadístico
str(iris)  # Estructura del dataset
```

# Estadística Descriptiva

Aquí calculamos medidas de tendencia central y dispersión para analizar nuestros datos.

```r
mean(iris$Sepal.Length)  # Media
median(iris$Sepal.Length)  # Mediana
```

```r
var(iris$Sepal.Length)  # Varianza
sd(iris$Sepal.Length)  # Desviación estándar
range(iris$Sepal.Length)  # Rango
```

# Distribuciones de probabilidad

Las distribuciones de probabilidad modelan el comportamiento de variables aleatorias.

Veamos ejemplos de la distribución normal, binomial y de Poisson.

```r
x <- seq(-4, 4, length=100)
y <- dnorm(x)
plot(x, y, type="l", main="Distribución Normal", ylab="Densidad", xlab="Valor")
```

```r
barplot(dbinom(0:10, size = 10, prob = 0.5), names.arg = 0:10, col = "blue",
        main = "Distribución Binomial", xlab = "Número de Éxitos", ylab = "Probabilidad")
```

```r
barplot(dpois(0:10, lambda = 3), names.arg = 0:10, col = "red",
        main = "Distribución de Poisson", xlab = "Número de Eventos", ylab = "Probabilidad")
```

# Visualización de datos

Los gráficos son esenciales para entender la distribución de los datos.
Aquí generamos histogramas y boxplots.

```r
ggplot(iris, aes(x = Sepal.Length)) +
  geom_histogram(binwidth = 0.2, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Distribución de Sepal Length", x = "Longitud del Sépalo", y = "Frecuencia")
```

```r
ggplot(iris, aes(x = Species, y = Sepal.Length, fill = Species)) +
  geom_boxplot() +
  labs(title = "Distribución de Sepal Length por Especie")
```
